```python
# Strings
# sWAP cASE
def swap_case(s):
    new_s=''
    for i in s:
        if i.isupper():
            new_s=new_s+ i.lower()
        else:
            new_s=new_s+ i.upper()

    return new_s
```

```python
# Find a string
def count_substring(string, sub_string):
    total=0
    for i in range(len(string)-len(sub_string)+1):
        sub_total=0
        for j in range(len(sub_string)):
            if string[j+i]==sub_string[j]:
                sub_total+=1
        if sub_total==len(sub_string):
            total +=1
    return total
```

```python
# Recursive Digit Sum
def superDigit(n, k):
    msum=0
    for i in n:
        msum +=int(i)
    msum *= int(k)
    if msum<=9:
        return msum
    else:
        return(superDigit(str(msum),"1"))
    

if __name__ == '__main__':

    n, k =map(str, input().split())
    print(superDigit(n,k))
```
